```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "../Resources/Composition of Debt Balance per Capita by State (2022 Q3).csv"
state_debt_df = pd.read_csv(csv_file)
state_debt_df
```

```python
state_debt = state_debt_df.rename(columns={'State':'state','Mortgage':'mortgage','HE Revolving':'home_equity',
                                         'Auto Loan':'auto_loan', 'Credit Card':'credit_card', 'Student Loan': 'student_loan',
                                         'Other':'other'})
state_debt
```

```python
csv_file = "../Resources/Debt Share by Product Type and Age (2022Q3).csv"
debt_age_df = pd.read_csv(csv_file)
debt_age_df.head()
```

```python
v1_debt_age_df = debt_age_df.drop(debt_age_df.index[8:28])
v1_debt_age_df
```

```python
v2_debt_age_df= v1_debt_age_df.drop(v1_debt_age_df.index[0:2])
v2_debt_age_df
```

```python
debt_age = v2_debt_age_df.rename(columns={'Debt Share by Product Type and Age (2022Q3)':'loan_type','Unnamed: 1':'18-29',
                                                  'Unnamed: 2':'30-39','Unnamed: 3':'40-49','Unnamed: 4':'50-59','Unnamed: 5':'60-69',
                                                 'Unnamed: 6':'70+'})

debt_age
```
