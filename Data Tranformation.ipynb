```python
#import csv file to clean Treasury site
csv_file = "Resources/DebtPenny_20050401_20221205.csv"
DebtPenny_data_df = pd.read_csv(csv_file)
DebtPenny_data_df.head()
```

```python
#Data Cleaning for Fred Household Debt
csv_file = "Resources/FREDHouseholdDebt.csv"
Fred_household_data_df = pd.read_csv(csv_file)
Fred_household_data_df.head()
```
